```python
class Board:
    def __init__(self):
        # Create an empty board represented by a list of 9 spaces
        self.cells = [' ' for _ in range(9)]

    def display(self):
        # Print the board to the terminal
        print("\n")
        print(f" {self.cells[0]} | {self.cells[1]} | {self.cells[2]} ")
        print("---+---+---")
        print(f" {self.cells[3]} | {self.cells[4]} | {self.cells[5]} ")
        print("---+---+---")
        print(f" {self.cells[6]} | {self.cells[7]} | {self.cells[8]} ")
        print("\n")

    def update_cell(self, position, player_symbol):
        # Check if the cell is empty before placing the symbol
        if self.cells[position] == ' ':
            self.cells[position] = player_symbol
            return True
        return False

    def check_winner(self, symbol):
        # All possible winning combinations (rows, columns, diagonals)
        winning_combinations = [
            [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
            [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
            [0, 4, 8], [2, 4, 6]              # Diagonals
        ]
        for combo in winning_combinations:
            if self.cells[combo[0]] == self.cells[combo[1]] == self.cells[combo[2]] == symbol:
                return True
        return False

    def is_full(self):
        # If there are no empty spaces left, the board is full (Draw)
        return ' ' not in self.cells


class Game:
    def __init__(self):
        self.board = Board()  # The game initializes the board
        self.current_player = 'X'

    def switch_player(self):
        # Switch turns between X and O
        self.current_player = 'O' if self.current_player == 'X' else 'X'

    def start(self):
        print("Welcome to Tic-Tac-Toe!")
        print("Positions are numbered 1-9 (from top-left to bottom-right).")

        while True:
            self.board.display()

            # Get user input and ensure it's an integer between 1 and 9
            try:
                move = int(input(f"Player ({self.current_player})'s turn - Choose a position (1-9): "))
                if move < 1 or move > 9:
                    print(" Invalid input. Please choose a number between 1 and 9.")
                    continue
            except ValueError:
                print(" Please enter a valid number, not text.")
                continue

            position = move - 1  # Subtract 1 because Python lists are 0-indexed

            # Update the board
            if self.board.update_cell(position, self.current_player):

                # Check for a win
                if self.board.check_winner(self.current_player):
                    self.board.display()
                    print(f" Congratulations! Player ({self.current_player}) wins!")
                    break

                # Check for a draw
                elif self.board.is_full():
                    self.board.display()
                    print(" It's a draw!")
                    break

                # If no win or draw, switch to the other player
                self.switch_player()
            else:
                print(" That cell is already taken! Choose another one.")


# Run the game
if __name__ == "__main__":
    tic_tac_toe = Game()
    tic_tac_toe.start()
```
