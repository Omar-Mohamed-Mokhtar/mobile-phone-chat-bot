```python
import re
import random

class MobileShopBot:

    negative_responses = ("no", "nope", "nah", "not interested")

    exit_commands = ("quit", "exit", "bye", "goodbye", "later")

    def __init__(self):
        self.mobile_responses = {
            'buy_phone': r'.*\b(buy|purchase|phone|mobile)\b.*',
            'price': r'.*\b(price|cost|how much)\b.*',
            'repair': r'.*\b(repair|fix|broken|screen)\b.*',
            'accessories': r'.*\b(case|charger|headphones|accessories)\b.*',
            'general_help': r'.*\b(help|assist)\b.*'
        }

    def greet(self):
        name=input("Enter your name:")
        print("Hello!",name, "Welcome to our Mobile Shop ")
        print("How can I help you today?")

    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply == command:
                print("Bot: Thank you for visiting our mobile shop. Goodbye!")
                return True

    def chat(self):
        self.greet()

        while True:
            user_input = input("You: ").lower()

            if self.make_exit(user_input):
                break

            self.respond(user_input)

    def respond(self, user_input):

        for intent, pattern in self.mobile_responses.items():

            if re.search(pattern, user_input):

                if intent == 'buy_phone':
                    print("Bot: We have Samsung, iPhone and Xiaomi phones available.")

                elif intent == 'price':
                    print("Bot: Prices start from 3000 EGP depending on the model.")

                elif intent == 'repair':
                    print("Bot: Yes, we repair screens, batteries and charging ports.")

                elif intent == 'accessories':
                    print("Bot: We sell chargers, cases, screen protectors and headphones.")

                elif intent == 'general_help':
                    print("Bot: I can help you buy a phone, check prices or repair phones.")

                return

        print("Bot: Sorry, I didn't understand. Can you ask in another way?")


bot = MobileShopBot()
bot.chat()
```
